# Hashtag Figure

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook
import couchdb
from collections import Counter
from bokeh.embed import components
from bokeh.resources import CDN
```

```python
url = "http://localhost:5984"
db_name = "test_db"
db = couchdb.Server(url=url)[db_name]
data = db.get('trending_hashtags')

count_terms_only = Counter()
count_terms_only.update(data['data'])
top10 = count_terms_only.most_common(10)
x, y = zip(*top10)
```

```python
f = figure(x_range=x,title="Hashtag Occurrences",
           toolbar_location='right', plot_height=399, plot_width=639, background_fill_color='gray',
          background_fill_alpha=0.1)
f.vbar(x=x, top=y, width=0.7)
f.y_range.start = 0
f.x_range.range_padding = 0.1
f.xaxis.major_label_orientation = -1
output_notebook()
show(f)
```

```python
script, div = components(f)
cdn_js = CDN.js_files
cdn_css =CDN.css_files
```
